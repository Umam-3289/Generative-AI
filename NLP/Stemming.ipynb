Stemming

Stemming is a text processing technique used in natural language processing (NLP) to reduce words to their root or base form, called a stem. It's commonly used to normalize text before analysis, searching, or machine learning.
How It Works
Stemming removes suffixes and prefixes from words using predefined rules. For example:

"running", "runs", "ran" → "run"
"happiness", "happily", "happy" → "happi"
"connection", "connecting", "connected" → "connect"

The resulting stem may not be a valid dictionary word (notice "happi" in the example above), which distinguishes stemming from lemmatization.
Common Stemming Algorithms
Porter Stemmer (most popular)

Rule-based approach developed in 1980
Works by removing common suffixes step-by-step
Widely used and relatively simple
Available in most NLP libraries

Snowball

More advanced framework that improves on Porter
Supports multiple languages
Better handling of edge cases

Lancaster Stemmer

More aggressive than Porter
Removes more characters, sometimes too aggressively

Lovins Stemmer

Single-pass algorithm
Less commonly used today

```python
# Example of porter stemmer
from nltk.stem import PorterStemmer
words=["happiness", "happily", "happy","connection", "connecting", "connected","running", "runs", "ran"]

print("The Result of Porter Stemmer --->\n")
stemming=PorterStemmer()
for word in words:
    print(word+"----->"+stemming.stem(word))
```

RegexpStemmer Class

RegexpStemmer is a flexible stemming class from NLTK that allows you to define custom stemming rules using regular expressions instead of relying on pre-built algorithms.

How It Works ??

The regular expression pattern matches suffixes (or other patterns) at the end of words. When a match is found, it gets replaced with nothing (removed), leaving the stem.

```python
# Example of RegexpStemmer class
from nltk.stem import RegexpStemmer

print("The Result of Regexp Stemmer --->\n")
reg_stemmer=RegexpStemmer(r'(ing|ness|ly|ion|ed|s)$')
for word in words:
    print(f"{word}-->{reg_stemmer.stem(word)}")
```

```python
#Example of snow ball stemmer 
from nltk.stem import SnowballStemmer

print("The Result of Snowball Stemmer --->\n")
Snow_ball_Stemmer=SnowballStemmer('english')
for word in words:
    print(f"{word}---->{Snow_ball_Stemmer.stem(word)}")
```

```python
# Example of Lancaster Stemmer
from nltk.stem import LancasterStemmer
lanc_stem=LancasterStemmer()

print("The Result of Lancaster Stemmer --->\n")
for word in words:
    print(f"{word}---->{lanc_stem.stem(word)}")
```

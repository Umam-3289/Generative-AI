Lemmatization

Lemmatization is a natural language processing (NLP) technique that reduces words to their base or dictionary form, called a lemma. It's a fundamental tool for text analysis and machine learning.
How It Works
Lemmatization analyzes the morphological structure of words and uses language rules to convert them to their base form. For example:

"running", "runs", "ran" → "run"
"better", "best" → "good"
"countries", "country's" → "country"
"studying", "studied" → "study"

Lemmatization vs. Stemming

Uses dictionary & grammar rulesUses  |  heuristic rules

Valid dictionary words (lemmas)  |  Rough word fragments

More accurate  |  Faster but less accurate

"walked" → "walk"  |  "walked" → "walk" / "wal"

Why it is Useful ?

Reduces dimensionality — Fewer unique tokens in text analysis

Improves search — Different forms of a word are treated as the same concept

Enhances machine learning — Helps models generalize better

Better text understanding — Reveals semantic relationships between words

Reduces noise — Filters out grammatical variations that don't change meaning


Common Use Cases

Search engines — Finding all documents about "running" when user searches "run"

Sentiment analysis — Recognizing "happy", "happier", "happily" as related concepts

Text classification — Grouping documents more effectively

Information retrieval — Improving document matching

Chatbots — Understanding user intent regardless of verb tense


In lemmatization "pos" tag plays very crucial role by applying pos type it converts the word into its base word

Types-->
    noun->'n'
    verb->'v'
    adjective->'a'
    adverb->'r'

```python
# Example of Lemmatization
# Noun Lemmatizer
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
noun=["children", "boxes", "geese", "studies", "analyses", "phenomena", "teeth", "potatoes"]

print("The Result after Noun lemmatization--->\n")
for word in noun:
    print(f"{word}---->{lemmatizer.lemmatize(word,pos='n')}")


```

```python
# Verb Lemmatizer

verb=["speaking", "written", "goes", "swimming", "bought", "breaking", "sleeping", "teaching"]

print("The Result after Verb lemmatization--->\n")
for word in verb:
    print(f"{word}---->{lemmatizer.lemmatize(word,pos='v')}")
```

```python
# Adjective Lemmatization

adjective=["better", "worst", "larger", "fastest", "happier", "more beautiful", "smaller", "greatest"]

print("The Result after Adjective lemmatization--->\n")
for word in adjective:
    print(f"{word}---->{lemmatizer.lemmatize(word,pos='a')}")
```

```python
# Adverb lemmatization

adverb=["quickly", "happily", "slowly", "carefully", "better", "worse", "faster", "more beautifully"]

print("The Result after Adverb lemmatization--->\n")
for word in adverb:
    print(f"{word}---->{lemmatizer.lemmatize(word,pos="r")}")
```

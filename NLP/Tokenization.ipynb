Tokenization in NLP

Tokenization is the first and fundamental step in NLP where you break down text into smaller units called tokens. A token can be a word, subword, or character, depending on the approach.
Why Tokenization Matters
Computers don't understand text the way humans do. Before any analysis can happen, text must be broken into manageable pieces that the model can process. Tokenization sets the foundation for everything that follows in NLP.

```python
corpus='''Hey Umam this side, I'm Tech enthusiasts.Currently seeking for Gen Ai Job Roles.
Let's connect together to cooperate each other in this learning journey.Always looking forward to 
contribute in real world applications.
'''

print(corpus)
```

```python
import nltk
# nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Toknization of Corpus to Documents
from nltk.tokenize import sent_tokenize


Documents=sent_tokenize(corpus,language="english")

print(Documents)

for sent in Documents:
    print(sent)
```

```python
# Tokenization of word from paragraph
from nltk.tokenize import word_tokenize

words=word_tokenize(corpus)
print(words)
```

```python
for sent in Documents:
    print(word_tokenize(sent))
```

```python
# Tokenization of corpus and treat punctuations as a token
from nltk.tokenize import wordpunct_tokenize
wt=wordpunct_tokenize(corpus)
print(wt)
```

```python
# TreeBankWord Tokenizer which treats only last punctuations of text as a separate token
from nltk.tokenize import TreebankWordTokenizer
tbt=TreebankWordTokenizer()
print(tbt.tokenize(corpus))
```
